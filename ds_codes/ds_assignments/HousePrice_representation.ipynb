```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset into a DataFrame
df = pd.read_csv('House Price India.csv')


# Create a histogram of the Price column
plt.hist(df['Price'], bins=50)
plt.title('Price Histogram')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Create a scatter plot of Price vs Living Area
plt.scatter(df['living area'], df['Price'])
plt.title('Price vs Living Area')
plt.xlabel('Living Area')
plt.ylabel('Price')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['Distance from the airport'], df['grade of the house'], df['Price'], s=df['living area']*0.01, alpha=0.5, c='royalblue')
ax.set_xlabel('Distance from the airport')
ax.set_ylabel('Grade of the house')
ax.set_zlabel('Price')
plt.show()
```
